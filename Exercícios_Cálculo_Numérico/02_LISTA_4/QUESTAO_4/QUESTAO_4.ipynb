```python
import numpy as np
import matplotlib.pyplot as plt
import math as mt
```

```python
# Funções e Análise

#Considerando X1 = x  e X2 = y, isolando o termo y em ambas as equações 
x = np.linspace(-20,11)
y_1 = np.array(9.5 - 0.5*x)
y_2 = np.array(9.4 - 0.51*x)
```

```python
#Plotagem do Gráfico
```

```python
plt.figure(figsize=(10,10))
plt.plot(x,y_1)
plt.plot(x,y_2)
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```
