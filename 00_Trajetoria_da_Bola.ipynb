Se assumirmos que a resistência do ar é desprezivel e ignorarmos a curvatura da Terra, uma bola lançada ao ar, de qualquer pornto da superficie da Terra, seguirá um pergucrso de voo parabólico, e a altura da bola em um instante t depois do lançamento é dada pela equação:

Equ.1.1

$$
\begin{aligned}
y(t) = y_{0} + \nu_{y0} + \frac{1}{2} \times g \times t^2 
\end{aligned}
$$

Onde o $$y_{0} $$ é a altura do objeo acima do solo, $$\nu_{y0}$$ é a velocidade vertical inicial do objeto, e o $$g$$, é a aceleração devido à gravidade da Terra. A distância horizontal percorrida pela bla como função do tempo após o lançamento é dada pela equação:

$$
\begin{aligned}
x(t) = x_{0} + \nu_{x0} \times t
\end{aligned}
$$

Onde o $$ x_{0}$$ é a posição horizontal inicial da bola no solo e $$\nu_{x0}$$ é a velocidade
horizontal da bola.

Se a boloa for lançada com uma velocidade inicial $$\nu_{0}$$ e um angulo $$\theta$$ graus
com relação a superficie da Terra, os componentes horizontal e vertical da velocidade serão:

$$
\begin{aligned}
\nu_{x0} = \nu_{0} \times \cos{\theta}
\end{aligned}
$$

$$
\begin{aligned}
\nu_{y0} = \nu_{0} \times \sin{\theta}
\end{aligned}
$$

Assumindo que a bola é lançada da posição $$(x_{0},y_{0}) = (0,0)$$ determine a trajetória da bola 
assumindo que ela é lançada com velocidade inicial de: $$\nu_{0} = 20 \, \frac{m}{s}$$

```python
import numpy as np
import math as mt
import pandas as pd
import matplotlib.pyplot as plt
g = 9.81
print("Ok")
```

```python
#Criando um vetor ângulo
angulo_1 = np.array([float(x) for x in range(0,91)])
```

```python
#Criando as Equações do Problemas - Range Maximo

v_0 = 20 # m/s
v_x0 = v_0 * np.cos(np.deg2rad(angulo_1)) # Vetor velocidade em x
v_y0 = v_0 * np.sin(np.deg2rad(angulo_1)) # Vetor velocidade em y
max_temp = ((-2) * v_y0)/g                # Tempo máximo de percurso
x = max_temp * v_x0 
df = pd.DataFrame(angulo_1,columns = ["Angulo"])
df["Tempo Máximo"] = (-1)*max_temp
df["VX"] = v_x0
df["VY"] = v_y0
df["X"] = x
print(df)
```
